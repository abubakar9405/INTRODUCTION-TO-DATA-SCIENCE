```python
#TASK-01 to import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
#TASK-02 to read file
df=pd.read_csv("iris.csv")
SepalLengthCm=df['SepalLengthCm']
SepalWidthCm=df['SepalWidthCm']
PetalLengthCm=df['PetalLengthCm']
spicies=df['Species']
#TASK-03
print(SepalLengthCm)
print("max value of sepal length column is:",SepalLengthCm.max())
print("min value of sepal length column is:",SepalLengthCm.min())
print("MODE:",SepalLengthCm.mode())
print("MEDIAN:",SepalLengthCm.median())
print("MEAN:",SepalLengthCm.mean())
print("STANDARD DEVIATION:",SepalLengthCm.std())
print("VARIANCE:",SepalLengthCm.var())
Q1=SepalLengthCm.quantile(0.25)
Q3=SepalLengthCm.quantile(0.75)
#calculating IQR
IQR=Q3-Q1
LOWER_BOUND=Q1-1.5*IQR
UPPER_BOUND=Q3+1.5*IQR
outliers=df[(df["SepalLengthCm"]<LOWER_BOUND) | (df["SepalLengthCm"]>UPPER_BOUND)]
print("OUTLIERS IN SEPAL LENGTH ARE:",outliers)
#TASK-04
print("HISTOGRAM OF SEPAL LENGTH")
plt.figure(figsize=(10,7))
plt.hist(SepalLengthCm,bins=20,color='red',edgecolor='black',alpha=0.7)
plt.xlabel('Sepal length (cm)')
plt.ylabel('frequency')
plt.grid(True)
plt.show()

print("HISTOGRAM FOR SEPAL WIDTH")
plt.figure(figsize=(10,7))
plt.hist(SepalWidthCm,bins=20,color='green',edgecolor='black',alpha=0.7)
plt.xlabel('sepal width(cm)')
plt.ylabel('frequency')
plt.grid(True)
plt.show()
print("HISTOGRAM FOR PETAL LENGTH")
plt.figure(figsize=(10,7))
plt.hist(PetalLengthCm,bins=20,color='yellow',edgecolor='black',alpha=0.7)
plt.grid(True)
plt.show()
print("BOX PLOT FOR SEPAL LENGTH")
data=[SepalLengthCm,SepalWidthCm,PetalLengthCm]
plt.figure(figsize=(10,7))
plt.boxplot(data,labels=['SepalLengthCm','SepalWidthCm','PetalLengthCm'])
plt.xlabel('sepal widht(cm)')
plt.ylabel('frequency')
plt.grid(False)
plt.show()
print("SCATTER PLOT B/W SEPAL LENGTH AND SEPAL WIDTH")
colors = {'Iris-setosa': 'blue', 'Iris-versicolor': 'green', 'Iris-virginica': 'red'}
plt.figure(figsize=(10, 7))

for sp in colors.keys():
    subset = df[df['Species'] == sp]
    plt.scatter(subset['SepalLengthCm'], subset['SepalWidthCm'],
                color=colors[sp], alpha=0.7, label=sp)


plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Scatter Plot of Sepal Length vs Sepal Width by Species')
plt.legend(title='Species')
plt.grid(True)

# Optionally add labels to points
for sp in colors.keys():
    subset = df[df['Species'] == sp]
    for i in range(len(subset)):
        plt.text(subset['SepalLengthCm'].iloc[i], subset['SepalWidthCm'].iloc[i],
                 subset['Species'].iloc[i], fontsize=9, alpha=0.7)

plt.show()
#TASK-05
species_count=df['Species'].value_counts()
print("unique species and their counts")
print(species_count)


```
